```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
from google.colab import drive
drive.mount('/content/drive')

data_2020 = pd.read_csv('/content/drive/MyDrive/Data/fatal_2020_medical.csv')
data_2020
```

```python
data2020 = data_2020.drop(columns = {'Total.2', 'Average'})
data2020
```

```python
modified_2020 = data2020.rename(columns={"Total": "Age", "Per Capita": 'Falls', 'Total.1': 'Total_Cost_M', 'Average.1': "Avg_Cost1_M", 'Per Capita.1': 'PC_1', 'Average.2': 'Avg_Cost2_B', 'Per Capita.2': 'PC_2_M' })
modified_2020
```

```python
# Remove "M", "$" and "B" from each column
modified_2020['Total_Cost_M'] = pd.to_numeric(modified_2020['Total_Cost_M'].astype(str).str.replace("[\$, M]", "", regex=True))
modified_2020['Avg_Cost1_M'] = pd.to_numeric(modified_2020['Avg_Cost1_M'].astype(str).str.replace("[\$, M]","", regex=True))
modified_2020['PC_1']= pd.to_numeric(modified_2020['PC_1'].astype(str).str.replace("$", "").str.replace(",", "")).astype(int)
modified_2020['Avg_Cost2_B'] = pd.to_numeric(modified_2020['Avg_Cost2_B'].astype(str).str.replace("[\$, B]", "", regex=True))
modified_2020['PC_2_M'] = pd.to_numeric(modified_2020['PC_2_M'].astype(str).str.replace("[\$, M]", "", regex=True))
modified_2020
```

```python
plt.plot(modified_2020['Age'], modified_2020['Falls'])
plt.xlabel('Age')
plt.ylabel('Falls')
plt.title('Falls in Elderly 2020')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot('Age', 'Falls', data=modified_2020, color='blue', label='Falls')
ax.set_xlabel("Age")
ax.set_ylabel("Falls")

ax2= ax.twinx()
ax2.plot("Age", "Total_Cost_M", data=modified_2020, color='red', label="Total Cost (Millions)")
ax2.set_ylabel("Total Cost (Millions)")
ax2.set_ylim([0, max(modified_2020["Total_Cost_M"]) +10])

plt.title("Number of Falls and Total Cost by Age")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot('Age', 'PC_1', data=modified_2020, color='red', label='Per Capita Cost by Age')
ax.set_xlabel("Age")
ax.set_ylabel("Per Capita Cost")

ax2= ax.twinx()
ax2.plot("Age", "Avg_Cost2_B", data=modified_2020, color='green', label="Average Cost (Billions)")
ax2.set_ylabel("Average Cost (Billions)")
ax2.set_ylim([0, max(modified_2020["Avg_Cost2_B"]) +5])

plt.title("Average Cost of Falls by Age")
plt.legend()
plt.show()
```
