```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
from google.colab import drive
drive.mount('/content/drive')

data_2018 = pd.read_csv('/content/drive/MyDrive/Data/fatal_2018_medical.csv')
data_2018
```

```python
data2018 = data_2018.drop(columns = {'Average', 'Total.1'})
data2018
```

```python
modified_2018 = data2018.rename(columns={"Per Capita": 'Falls', 'Average.1': "Avg_Cost1_M", 'Per Capita.1': 'PC_1', 'Average.2': 'Avg_Cost2_B', 'Per Capita.2': 'PC_2_M' })
modified_2018
```

```python
# Remove "M", "$" and "B" from each column
modified_2018['Total'] = pd.to_numeric(modified_2018['Total'].astype(str).str.replace("[\$, M]", "", regex=True))
modified_2018['Avg_Cost1_M'] = pd.to_numeric(modified_2018['Avg_Cost1_M'].astype(str).str.replace("[\$, M]","", regex=True))
modified_2018['PC_1']= pd.to_numeric(modified_2018['PC_1'].astype(str).str.replace("$", "").str.replace(",", "")).astype(int)
modified_2018['Avg_Cost2_B'] = pd.to_numeric(modified_2018['Avg_Cost2_B'].astype(str).str.replace("[\$, B]", "", regex=True))
modified_2018['PC_2_M'] = pd.to_numeric(modified_2018['PC_2_M'].astype(str).str.replace("[\$, M]", "", regex=True))
modified_2018
```

```python
plt.plot(modified_2018['Age'], modified_2018['Falls'])
plt.xlabel('Age')
plt.ylabel('Falls')
plt.title('Falls in Elderly 2018')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot('Age', 'Falls', data=modified_2018, color='blue', label='Falls')
ax.set_xlabel("Age")
ax.set_ylabel("Falls")

ax2= ax.twinx()
ax2.plot("Age", "Avg_Cost1_M", data=modified_2018, color='red', label="Average Cost (Millions)")
ax2.set_ylabel("Average Cost (Millions)")
ax2.set_ylim([0, max(modified_2018["Avg_Cost1_M"]) +10])

plt.title("Number of Falls and Total Cost by Age")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot('Age', 'PC_1', data=modified_2018, color='red', label='Per Capita Cost by Age')
ax.set_xlabel("Age")
ax.set_ylabel("Per Capita Cost")

ax2= ax.twinx()
ax2.plot("Age", "Avg_Cost2_B", data=modified_2018, color='green', label="Average Cost (Billions)")
ax2.set_ylabel("Average Cost (Billions)")
ax2.set_ylim([0, max(modified_2018["Avg_Cost2_B"]) +5])

plt.title("Average Cost of Falls by Age")
plt.legend()
plt.show()
```
